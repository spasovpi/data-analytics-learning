```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/Zara_Sales_Analysis.csv", delimiter=';', encoding='latin1')
print("✅ Файл успешно загружен!")
```

```python
print("=== ИНФОРМАЦИЯ О ДАТАСЕТЕ ===")
print(f"Размер датасета: {df.shape}")  # (количество строк, количество колонок)
print(f"Количество строк: {df.shape[0]}")
print(f"Количество колонок: {df.shape[1]}")

print("\n=== ПЕРВЫЕ 5 СТРОК ДАТАСЕТА ===")
df.head()
```

```python
# Посмотрим на имена всех колонок
print("=== СПИСОК ВСЕХ КОЛОНОК ===")
print(df.columns.tolist())

# Посмотрим на основные статистические показатели числовых колонок
print("\n=== ОСНОВНАЯ СТАТИСТИКА ===")
print(df.describe())

# Посмотрим на типы данных в каждой колонке
print("\n=== ТИПЫ ДАННЫХ ===")
print(df.dtypes)

# Посмотрим на последние 5 строк для полноты картины
print("\n=== ПОСЛЕДНИЕ 5 СТРОК ===")
df.tail()
```

```python
# Проверим пропущенные значения
print("=== ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ ===")
print(df.isnull().sum())

# Посмотрим на уникальные значения в ключевых категориальных колонках
print("\n=== УНИКАЛЬНЫЕ КАТЕГОРИИ ===")
print("Product Category:", df['Product Category'].unique())
print("Product Position:", df['Product Position'].unique())
print("Promotion:", df['Promotion'].unique())
print("Seasonal:", df['Seasonal'].unique())

# Проверим проблемные значения в колонке price
print("\n=== ПРОБЛЕМНЫЕ ЗНАЧЕНИЯ В ЦЕНАХ ===")
print("Уникальные значения в currency:", df['currency'].unique())
print("Примеры проблемных цен:")
problem_prices = df[~df['price'].apply(lambda x: str(x).replace('.', '').isdigit())]
print(problem_prices[['name', 'price', 'currency']].head(10))
```

```python
# Создадим копию датафрейма для безопасной работы
df_clean = df.copy()

print("=== ОЧИСТКА ДАННЫХ ===")

# 1. Удалим строки с пропущенными названиями товаров
initial_count = len(df_clean)
df_clean = df_clean.dropna(subset=['name'])
print(f"Удалено строк с пропущенными названиями: {initial_count - len(df_clean)}")

# 2. Почистим колонку price - преобразуем в числовой формат
# Сначала заменим проблемные символы
df_clean['price_clean'] = df_clean['price'].astype(str).str.replace('[^0-9.]', '', regex=True)

# Преобразуем в числовой формат
df_clean['price_clean'] = pd.to_numeric(df_clean['price_clean'], errors='coerce')

# Проверим результат
print(f"Проблемных цен после очистки: {df_clean['price_clean'].isnull().sum()}")

# 3. Проверим основные статистики по очищенным ценам
print("\n=== СТАТИСТИКА ПО ОЧИЩЕННЫМ ЦЕНАМ ===")
print(df_clean['price_clean'].describe())

# 4. Посмотрим на распределение категорий
print("\n=== РАСПРЕДЕЛЕНИЕ ПО КАТЕГОРИЯМ ===")
category_counts = df_clean['Product Category'].value_counts()
print(category_counts)

# 5. Посмотрим на распределение по разделам (section)
print("\n=== РАСПРЕДЕЛЕНИЕ ПО РАЗДЕЛАМ ===")
section_counts = df_clean['section'].value_counts()
print(section_counts)
```

```python
# Преобразуем колонку scraped_at в формат даты
print("=== ПРЕОБРАЗОВАНИЕ ДАТ ===")
df_clean['scraped_at'] = pd.to_datetime(df_clean['scraped_at'])

# Извлечем компоненты даты для анализа
df_clean['date'] = df_clean['scraped_at'].dt.date
df_clean['month'] = df_clean['scraped_at'].dt.month
df_clean['year'] = df_clean['scraped_at'].dt.year
df_clean['day'] = df_clean['scraped_at'].dt.day

# Создадим функцию для определения сезона
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df_clean['season'] = df_clean['month'].apply(get_season)

# Проверим результат
print("Добавленные колонки:")
print(df_clean[['scraped_at', 'date', 'month', 'year', 'season']].head())

print("\n=== ИНФОРМАЦИЯ О ДАТАХ ===")
print(f"Диапазон дат: от {df_clean['date'].min()} до {df_clean['date'].max()}")
print(f"Уникальные месяцы: {sorted(df_clean['month'].unique())}")
print(f"Распределение по сезонам:")
print(df_clean['season'].value_counts())

# Посмотрим на общую структуру данных после преобразований
print("\n=== ОБНОВЛЕННАЯ СТРУКТУРА ДАННЫХ ===")
print(f"Размер датасета: {df_clean.shape}")
print(f"Колонки: {df_clean.columns.tolist()}")
```

```python
print("=== ТОП-10 САМЫХ ПРОДАВАЕМЫХ ТОВАРОВ ===")

# Группируем по названию товара и агрегируем данные
top_products = df_clean.groupby('name').agg({
    'Sales Volume': 'sum',
    'price_clean': 'mean',
    'section': 'first',
    'terms': 'first'
}).sort_values('Sales Volume', ascending=False).head(10)

# Переименуем колонки для понятности
top_products = top_products.rename(columns={
    'Sales Volume': 'Total_Sales',
    'price_clean': 'Avg_Price',
    'section': 'Section',
    'terms': 'Category'
})

print("Топ-10 товаров по объему продаж:")
print(top_products[['Total_Sales', 'Avg_Price', 'Section']])

# Визуализируем топ-10 товаров
plt.figure(figsize=(12, 8))
bars = plt.barh(top_products.index, top_products['Total_Sales'], color='skyblue')
plt.xlabel('Объем продаж')
plt.title('ТОП-10 САМЫХ ПРОДАВАЕМЫХ ТОВАРОВ')
plt.gca().invert_yaxis()  # Чтобы самый продаваемый был сверху

# Добавим подписи с ценами на график
for i, bar in enumerate(bars):
    width = bar.get_width()
    price = top_products['Avg_Price'].iloc[i]
    plt.text(width + 20, bar.get_y() + bar.get_height()/2, 
             f'${price:.1f}', ha='left', va='center')

plt.tight_layout()
plt.show()

# Дополнительная статистика по топ-товарам
print(f"\n=== СТАТИСТИКА ПО ТОП-10 ТОВАРАМ ===")
print(f"Средняя цена в топ-10: ${top_products['Avg_Price'].mean():.2f}")
print(f"Медианная цена в топ-10: ${top_products['Avg_Price'].median():.2f}")
print(f"Средний объем продаж в топ-10: {top_products['Total_Sales'].mean():.0f} единиц")
```

```python
print("=== СРАВНЕНИЕ РАЗДЕЛОВ MAN vs WOMAN ===")

# Анализ по разделам
section_analysis = df_clean.groupby('section').agg({
    'Sales Volume': ['sum', 'mean', 'count'],
    'price_clean': ['mean', 'median', 'min', 'max']
}).round(2)

# Упростим названия колонок
section_analysis.columns = ['Total_Sales', 'Avg_Sales_Per_Product', 'Product_Count', 
                           'Avg_Price', 'Median_Price', 'Min_Price', 'Max_Price']

print("Сравнение разделов:")
print(section_analysis)

# Визуализация сравнения
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# График 1: Общий объем продаж по разделам
axes[0,0].bar(section_analysis.index, section_analysis['Total_Sales'], color=['blue', 'pink'])
axes[0,0].set_title('ОБЩИЙ ОБЪЕМ ПРОДАЖ ПО РАЗДЕЛАМ')
axes[0,0].set_ylabel('Объем продаж')

# График 2: Средняя цена по разделам
axes[0,1].bar(section_analysis.index, section_analysis['Avg_Price'], color=['lightblue', 'lightpink'])
axes[0,1].set_title('СРЕДНЯЯ ЦЕНА ПО РАЗДЕЛАМ')
axes[0,1].set_ylabel('Цена ($)')

# График 3: Количество товаров по разделам
axes[1,0].bar(section_analysis.index, section_analysis['Product_Count'], color=['darkblue', 'deeppink'])
axes[1,0].set_title('КОЛИЧЕСТВО ТОВАРОВ ПО РАЗДЕЛАМ')
axes[1,0].set_ylabel('Количество товаров')

# График 4: Средние продажи на товар
axes[1,1].bar(section_analysis.index, section_analysis['Avg_Sales_Per_Product'], color=['cornflowerblue', 'hotpink'])
axes[1,1].set_title('СРЕДНИЕ ПРОДАЖИ НА ТОВАР')
axes[1,1].set_ylabel('Продажи на товар')

plt.tight_layout()
plt.show()

# Дополнительные расчеты
print(f"\n=== ДОПОЛНИТЕЛЬНАЯ СТАТИСТИКА ===")
man_sales_ratio = section_analysis.loc['MAN', 'Total_Sales'] / section_analysis['Total_Sales'].sum() * 100
woman_sales_ratio = section_analysis.loc['WOMAN', 'Total_Sales'] / section_analysis['Total_Sales'].sum() * 100

print(f"Доля продаж MAN: {man_sales_ratio:.1f}%")
print(f"Доля продаж WOMAN: {woman_sales_ratio:.1f}%")
print(f"Соотношение товаров MAN:WOMAN = {section_analysis.loc['MAN', 'Product_Count']:.0f}:{section_analysis.loc['WOMAN', 'Product_Count']:.0f}")
```

```python
print("=== АНАЛИЗ ЗАВИСИМОСТИ ЦЕНЫ И ОБЪЕМА ПРОДАЖ ===")

# Рассчитаем корреляцию между ценой и объемом продаж
correlation = df_clean['price_clean'].corr(df_clean['Sales Volume'])
print(f"Корреляция между ценой и объемом продаж: {correlation:.3f}")

# Создадим ценовые сегменты для анализа
def price_segment(price):
    if price < 50:
        return '1_Бюджетный (<$50)'
    elif price < 100:
        return '2_Средний ($50-$100)'
    elif price < 200:
        return '3_Премиум ($100-$200)'
    else:
        return '4_Люкс (>$200)'

df_clean['price_segment'] = df_clean['price_clean'].apply(price_segment)

# Анализ по ценовым сегментам
segment_analysis = df_clean.groupby('price_segment').agg({
    'Sales Volume': ['sum', 'mean', 'count'],
    'price_clean': 'mean'
}).round(2)

segment_analysis.columns = ['Total_Sales', 'Avg_Sales_Per_Product', 'Product_Count', 'Avg_Price']

# Упорядочим сегменты в логическом порядке
segment_order = ['1_Бюджетный (<$50)', '2_Средний ($50-$100)', '3_Премиум ($100-$200)', '4_Люкс (>$200)']
segment_analysis = segment_analysis.reindex(segment_order)

# Уберем префиксы для красивого отображения
segment_analysis.index = segment_analysis.index.str.replace(r'^\d_', '', regex=True)

print("\nАНАЛИЗ ПО ЦЕНОВЫМ СЕГМЕНТАМ (от бюджетного к люксу):")
print(segment_analysis)

# Визуализация
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# График 1: Зависимость цены и объема продаж (scatter plot)
axes[0,0].scatter(df_clean['price_clean'], df_clean['Sales Volume'], alpha=0.6, color='purple')
axes[0,0].set_xlabel('Цена ($)')
axes[0,0].set_ylabel('Объем продаж')
axes[0,0].set_title('ЗАВИСИМОСТЬ: ЦЕНА vs ОБЪЕМ ПРОДАЖ')
axes[0,0].grid(True, alpha=0.3)

# График 2: Распределение по ценовым сегментам
segment_analysis['Product_Count'].plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('КОЛИЧЕСТВО ТОВАРОВ ПО ЦЕНОВЫМ СЕГМЕНТАМ')
axes[0,1].set_ylabel('Количество товаров')
axes[0,1].tick_params(axis='x', rotation=45)

# График 3: Объем продаж по ценовым сегментам
segment_analysis['Total_Sales'].plot(kind='bar', ax=axes[1,0], color='orange')
axes[1,0].set_title('ОБЩИЙ ОБЪЕМ ПРОДАЖ ПО ЦЕНОВЫМ СЕГМЕНТАМ')
axes[1,0].set_ylabel('Объем продаж')
axes[1,0].tick_params(axis='x', rotation=45)

# График 4: Средние продажи на товар по сегментам
segment_analysis['Avg_Sales_Per_Product'].plot(kind='bar', ax=axes[1,1], color='red')
axes[1,1].set_title('СРЕДНИЕ ПРОДАЖИ НА ТОВАР ПО СЕГМЕНТАМ')
axes[1,1].set_ylabel('Средние продажи на товар')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Дополнительный анализ - эффективность каждого сегмента
print(f"\n=== ЭФФЕКТИВНОСТЬ ЦЕНОВЫХ СЕГМЕНТОВ ===")
for segment in segment_analysis.index:
    data = segment_analysis.loc[segment]
    efficiency = data['Avg_Sales_Per_Product'] / data['Avg_Price']
    print(f"{segment}: {efficiency:.3f} продаж на $1 цены")

# Анализ эффективности промо-акций
print(f"\n=== АНАЛИЗ ПРОМО-АКЦИЙ ===")
promo_analysis = df_clean.groupby('Promotion').agg({
    'Sales Volume': ['sum', 'mean', 'count'],
    'price_clean': 'mean'
}).round(2)

promo_analysis.columns = ['Total_Sales', 'Avg_Sales_Per_Product', 'Product_Count', 'Avg_Price']
print("Эффективность товаров с промо-акциями и без:")
print(promo_analysis)
```

```python
# Давайте посчитаем эффективность более детально
print("=== ДЕТАЛЬНЫЙ АНАЛИЗ ЭФФЕКТИВНОСТИ ПРОМО-АКЦИЙ ===")

# Эффективность продаж на $1 цены
no_promo_efficiency = promo_analysis.loc['No', 'Avg_Sales_Per_Product'] / promo_analysis.loc['No', 'Avg_Price']
yes_promo_efficiency = promo_analysis.loc['Yes', 'Avg_Sales_Per_Product'] / promo_analysis.loc['Yes', 'Avg_Price']

print(f"Эффективность товаров БЕЗ промо: {no_promo_efficiency:.2f} продаж на $1 цены")
print(f"Эффективность товаров С промо: {yes_promo_efficiency:.2f} продаж на $1 цены")

# Разница в процентах
sales_diff_percent = ((promo_analysis.loc['Yes', 'Avg_Sales_Per_Product'] - promo_analysis.loc['No', 'Avg_Sales_Per_Product']) / 
                     promo_analysis.loc['No', 'Avg_Sales_Per_Product'] * 100)
price_diff_percent = ((promo_analysis.loc['Yes', 'Avg_Price'] - promo_analysis.loc['No', 'Avg_Price']) / 
                     promo_analysis.loc['No', 'Avg_Price'] * 100)

print(f"\nРазница в продажах: +{sales_diff_percent:.1f}%")
print(f"Разница в цене: +{price_diff_percent:.1f}%")
```

```python
print("=== АНАЛИЗ ЭФФЕКТИВНОСТИ РАСПОЛОЖЕНИЯ ТОВАРОВ ===")

# Анализ по расположению товаров
position_analysis = df_clean.groupby('Product Position').agg({
    'Sales Volume': ['sum', 'mean', 'count'],
    'price_clean': 'mean'
}).round(2)

position_analysis.columns = ['Total_Sales', 'Avg_Sales_Per_Product', 'Product_Count', 'Avg_Price']
position_analysis = position_analysis.sort_values('Total_Sales', ascending=False)

print("Эффективность различных расположений товаров:")
print(position_analysis)

# Визуализация расположения товаров
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# График 1: Общий объем продаж по расположению
position_analysis['Total_Sales'].plot(kind='bar', ax=axes[0,0], color='lightblue')
axes[0,0].set_title('ОБЩИЙ ОБЪЕМ ПРОДАЖ ПО РАСПОЛОЖЕНИЮ')
axes[0,0].set_ylabel('Объем продаж')
axes[0,0].tick_params(axis='x', rotation=45)

# График 2: Средние продажи на товар по расположению
position_analysis['Avg_Sales_Per_Product'].plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('СРЕДНИЕ ПРОДАЖИ НА ТОВАР ПО РАСПОЛОЖЕНИЮ')
axes[0,1].set_ylabel('Средние продажи')
axes[0,1].tick_params(axis='x', rotation=45)

# График 3: Количество товаров по расположению
position_analysis['Product_Count'].plot(kind='bar', ax=axes[1,0], color='orange')
axes[1,0].set_title('КОЛИЧЕСТВО ТОВАРОВ ПО РАСПОЛОЖЕНИЮ')
axes[1,0].set_ylabel('Количество товаров')
axes[1,0].tick_params(axis='x', rotation=45)

# График 4: Средняя цена по расположению
position_analysis['Avg_Price'].plot(kind='bar', ax=axes[1,1], color='pink')
axes[1,1].set_title('СРЕДНЯЯ ЦЕНА ПО РАСПОЛОЖЕНИЮ')
axes[1,1].set_ylabel('Средняя цена ($)')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Финальные выводы
print(f"\n=== ФИНАЛЬНЫЕ ВЫВОДЫ И РЕКОМЕНДАЦИИ ===")
print("📈 САМЫЕ ПРОДАВАЕМЫЕ КАТЕГОРИИ:")
top_categories = df_clean.groupby('terms')['Sales Volume'].sum().sort_values(ascending=False).head(5)
for i, (category, sales) in enumerate(top_categories.items(), 1):
    print(f"{i}. {category}: {sales:,} продаж")

print(f"\n💰 ЦЕНОВАЯ СТРАТЕГИЯ:")
best_segment = segment_analysis.loc[segment_analysis['Avg_Sales_Per_Product'].idxmax()]
print(f"Самый эффективный ценовой сегмент: {best_segment.name}")
print(f"Средние продажи: {best_segment['Avg_Sales_Per_Product']:.0f} на товар")
print(f"Средняя цена: ${best_segment['Avg_Price']:.2f}")

print(f"\n🏪 РЕКОМЕНДАЦИИ ПО РАСПОЛОЖЕНИЮ:")
best_position = position_analysis.loc[position_analysis['Avg_Sales_Per_Product'].idxmax()]
print(f"Самое эффективное расположение: {best_position.name}")
print(f"Средние продажи: {best_position['Avg_Sales_Per_Product']:.0f} на товар")
```
