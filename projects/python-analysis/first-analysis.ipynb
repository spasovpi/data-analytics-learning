```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry", 
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)
```

```python
# Основная информация о данных
df.info()

```

```python
# Статистика по возрасту
print(f"• Средний возраст: {round(df['Age'].mean())} лет")
print(f"• Минимальный возраст: {df['Age'].min()} лет") 
print(f"• Максимальный возраст: {df['Age'].max()} лет")
print(f"• Общее количество людей: {len(df)}")
```

```python
# Количество мужчин и женщин
male_count = df['Sex'].value_counts()['male']
female_count = df['Sex'].value_counts()['female']
print(f"• Мужчин: {male_count}")
print(f"• Женщин: {female_count}")
```

```python
# Более детальный анализ по полу
avg_age_male = df[df['Sex'] == 'male']['Age'].mean()
avg_age_female = df[df['Sex'] == 'female']['Age'].mean()
print(f"• Мужчины: {avg_age_male} лет")
print(f"• Женщины: {avg_age_female} лет")
```
